```python
df.columns = ['날짜', '종가', '거래량', '시가', '최고가', '최저가']
```

```python
df
```

```python
df['종가'] = df['종가'].str.replace('$','').astype(float)
df['시가'] = df['시가'].str.replace('$', '').astype(float)
df['최고가'] = df['최고가'].str.replace('$', '').astype(float)
df['최저가'] = df['최저가'].str.replace('$', '').astype(float)
```

```python
df.to_csv('./data/TSLA.csv')
```
